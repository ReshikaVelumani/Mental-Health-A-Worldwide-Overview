```python
import plotly.plotly as py
import plotly.graph_objs as go

import pandas as pd
import math

df = pd.read_csv("parallel_try1.csv")
df = df.sort_values(['Continent', 'Country'])
slope = 2.666051223553066e05
hover_text = []
bubble_size = []

for index, row in df.iterrows():
    hover_text.append(('Country: {country}<br>'+
                      'Depression %: {dep}<br>'+
                      'Males %: {males}<br>'+
                      'Females %: {females}<br>').format(country=row['Country'],
                                            males=row['Males %'],
                                            females=row['Females %'],
                                            dep=row['Depression %']))
    bubble_size.append(math.sqrt(row['GDP']*slope))

df['text'] = hover_text
df['size'] = bubble_size
sizeref = 5.*max(df['size'])/(100**2)

trace0 = go.Scatter(
    x=df['Depression %'][df['Continent'] == 'Africa'],
    y=df['Substance Misuse per 1M'][df['Continent'] == 'Africa'],
    mode='markers',
    name='Africa',
    text=df['text'][df['Continent'] == 'Africa'],
    marker=dict(
        symbol='circle',
        sizemode='area',
        sizeref=sizeref,
        size=df['size'][df['Continent'] == 'Africa'],
        line=dict(
            width=2
        ),
    )
)
trace1 = go.Scatter(
    x=df['Depression %'][df['Continent'] == 'Asia'],
    y=df['Substance Misuse per 1M'][df['Continent'] == 'Asia'],
    mode='markers',
    name='Asia',
    text=df['text'][df['Continent'] == 'Asia'],
    marker=dict(
        symbol='circle',
        sizemode='area',
        sizeref=sizeref,
        size=df['size'][df['Continent'] == 'Asia'],
        line=dict(
            width=2
        ),
    )
)
trace2 = go.Scatter(
    x=df['Depression %'][df['Continent'] == 'North America'],
    y=df['Substance Misuse per 1M'][df['Continent'] == 'North America'],
    mode='markers',
    name='North America',
    text=df['text'][df['Continent'] == 'North America'],
    marker=dict(
        symbol='circle',
        sizemode='area',
        sizeref=sizeref,
        size=df['size'][df['Continent'] == 'North America'],
        line=dict(
            width=2
        ),
    )
)
trace3 = go.Scatter(
    x=df['Depression %'][df['Continent'] == 'South America'],
    y=df['Substance Misuse per 1M'][df['Continent'] == 'South America'],
    mode='markers',
    name='South America',
    text=df['text'][df['Continent'] == 'South America'],
    marker=dict(
        symbol='circle',
        sizemode='area',
        sizeref=sizeref,
        size=df['size'][df['Continent'] == 'South America'],
        line=dict(
            width=2
        ),
    )
)
trace4 = go.Scatter(
    x=df['Depression %'][df['Continent'] == 'Australia'],
    y=df['Substance Misuse per 1M'][df['Continent'] == 'Australia'],
    mode='markers',
    name='Australia',
    text=df['text'][df['Continent'] == 'Australia'],
    marker=dict(
        symbol='circle',
        sizemode='area',
        sizeref=sizeref,
        size=df['size'][df['Continent'] == 'Australia'],
        line=dict(
            width=2
        ),
    )
)
trace5 = go.Scatter(
    x=df['Depression %'][df['Continent'] == 'Europe'],
    y=df['Substance Misuse per 1M'][df['Continent'] == 'Europe'],
    mode='markers',
    name='Europe',
    text=df['text'][df['Continent'] == 'Europe'],
    marker=dict(
        symbol='circle',
        sizemode='area',
        sizeref=sizeref,
        size=df['size'][df['Continent'] == 'Europe'],
        line=dict(
            width=2
        ),
    )
)
trace6 = go.Scatter(
    x=df['Depression %'][df['Continent'] == 'Oceania'],
    y=df['Substance Misuse per 1M'][df['Continent'] == 'Oceania'],
    mode='markers',
    name='Oceania',
    text=df['text'][df['Continent'] == 'Oceania'],
    marker=dict(
        symbol='circle',
        sizemode='area',
        sizeref=sizeref,
        size=df['size'][df['Continent'] == 'Oceania'],
        line=dict(
            width=2
        ),
    )
)

data = [trace0, trace1, trace2, trace3, trace4, trace5, trace6]
layout = go.Layout(
    title='Substance Misuse VS Percentage Depression, 2017',
    xaxis=dict(
        title='Depression Percentage',
        gridcolor='rgb(255, 255, 255)',
        range=[2, 7],
        zerolinewidth=1,
        ticklen=5,
        gridwidth=2,
    ),
    yaxis=dict(
        title='Substance Misuse per 1 Million',
        gridcolor='rgb(255, 255, 255)',
        range=[11000, 22000],
        zerolinewidth=1,
        ticklen=5,
        gridwidth=2,
    ),showlegend=True
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='substance_vs_depression')
```
